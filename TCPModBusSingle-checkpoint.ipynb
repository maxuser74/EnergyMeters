```python
from pymodbus.constants import Endian
from pymodbus.client import ModbusTcpClient
import struct
```

```python
cabina = '192.168.156.75'
```

```python
nodo = 8
```

```python
registro = 372 # Tensione RMS media su 3 fasi V
```

```python
registro = 374 # Corrente di linea L1 A
```

```python
registro = 376 # Corrente di linea L2 A
```

```python
registro = 378 # Corrente di linea L3 A
```

```python
registro = 390 # Potenza ATTIVA istantanea media RMS Watt
```

```python
client = ModbusTcpClient(cabina, port=502)
client.connect()

request = client.read_holding_registers(address=registro, count=2, slave=nodo)

# Convert the two 16-bit registers to a 32-bit float (avoiding deprecated BinaryPayloadDecoder)
high_word = request.registers[0]
low_word = request.registers[1]

# Convert to 32-bit float: word order is little endian (low word first)
# but byte order within each word is big endian
packed_data = struct.pack('>HH', low_word, high_word)  # Little word order, big endian bytes
valore = struct.unpack('>f', packed_data)[0]  # Big endian 32-bit float

print('Lettura cabina ',cabina,' - unità = ',nodo,' - Registro =',registro,' valore = ', valore)

client.close()
```
