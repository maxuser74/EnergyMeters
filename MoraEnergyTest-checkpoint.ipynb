```python
import pandas as pd

from pymodbus.constants import Endian
from pymodbus.payload import BinaryPayloadDecoder
from pymodbus.client.sync import ModbusTcpClient

import pandas.io.formats.excel
pandas.io.formats.excel.header_style = None

import os
import datetime
from datetime import datetime
```

```python
def validator(instance):
    if not instance.isError():
        '''.isError() implemented in pymodbus 1.4.0 and above.'''
        decoder = BinaryPayloadDecoder.fromRegisters(
            instance.registers,
            byteorder=Endian.Big, wordorder=Endian.Little
        )   
        return float('{0:.2f}'.format(decoder.decode_32bit_float()))

    else:
        # Error handling.
        print("There isn't the registers, Try again.")
        return None
        
richieste = [[1,372,'Tensione RMS media su 3 fasi totale cabina V = '],
             [1,374,'Corrente di linea totale cabina L1 A = '],
             [1,376,'Corrente di linea totale cabina L2 A = '],
             [1,378,'Corrente di linea totale cabina L3 A = '],
            [1,390,'Potenza ATTIVA istantanea media RMS totale cabina W = ']]
moduli = [[1,'192.168.156.75'], [2,'192.168.156.76'],[3,'192.168.156.77']]
```

```python
consumo_istantaneo_cabine = 0


# Cabina 1, unità 1 e 13
client = ModbusTcpClient('192.168.156.75', port=502)
request = client.read_holding_registers(390, 2, unit=1)
decoder = BinaryPayloadDecoder.fromRegisters(request.registers,byteorder=Endian.Big, wordorder=Endian.Little)           
valore = decoder.decode_32bit_float()
print('Consumo istantaneo cabina 1 - unità 1 W = ',valore)
consumo_istantaneo_cabine += valore

request = client.read_holding_registers(390, 2, unit=13)
decoder = BinaryPayloadDecoder.fromRegisters(request.registers,byteorder=Endian.Big, wordorder=Endian.Little)           
valore = decoder.decode_32bit_float()
print('Consumo istantaneo cabina 1 - unità 13 W = ',valore)
print()
consumo_istantaneo_cabine += valore
client.close()

# Cabina 2, unità 1
client = ModbusTcpClient('192.168.156.76', port=502)
request = client.read_holding_registers(390, 2, unit=1)
decoder = BinaryPayloadDecoder.fromRegisters(request.registers,byteorder=Endian.Big, wordorder=Endian.Little)           
valore = decoder.decode_32bit_float()
print('Consumo istantaneo cabina 2 - unità 1 W = ',valore)
consumo_istantaneo_cabine += valore

request = client.read_holding_registers(390, 2, unit=14)
decoder = BinaryPayloadDecoder.fromRegisters(request.registers,byteorder=Endian.Big, wordorder=Endian.Little)           
valore = decoder.decode_32bit_float()
print('Consumo istantaneo cabina 2 - unità 14 W = ',valore)
print()
consumo_istantaneo_cabine += valore

client.close()

# Cabina 3, unità 1
client = ModbusTcpClient('192.168.156.77', port=502)
request = client.read_holding_registers(390, 2, unit=1)
decoder = BinaryPayloadDecoder.fromRegisters(request.registers,byteorder=Endian.Big, wordorder=Endian.Little)           
valore = decoder.decode_32bit_float()
consumo_istantaneo_cabine += valore

print('Consumo istantaneo cabina 3 - unità 1 W = ',valore)
print('---------------------------------------------------')

client.close()

print('Consumo totale istantaneo - W = ',float('{0:.2f}'.format(consumo_istantaneo_cabine)))
```

```python
data = pd.DataFrame(columns = ['Cabina', 'Nodo', 'Valore'
#                               ,'Energy Unit Factor - 1=W / 4=Kwh',
#                               'CT_Transducer Ratio', 'Power <Factor',
#                               ,'L1','L2','L3'
                              ])

cabine = [[1,'192.168.156.75',27],[2,'192.168.156.76',21],[3,'192.168.156.77',16]]
print('Campionamento')

for i in cabine:
    print('Cabina ',i[0],end = '')
    client = ModbusTcpClient(i[1], port=502)
    for x in range(1,i[2]+1):
        #Lettura
        request = client.read_holding_registers(390, 2, unit=x)
        decoder = BinaryPayloadDecoder.fromRegisters(request.registers,byteorder=Endian.Big, wordorder=Endian.Little)           
        valore = decoder.decode_32bit_float()
        
        data.loc[len(data)+1] = [str(int(i[0])),
                                 str(int(x)),
                                 float("{0:.2f}".format(valore/1000))
                                 #,valore2
                                 #,valore3
                                 #,valore4
#                                 ,valore5
#                                ,valore6
#                                ,valore7
                                ]
        print('.', end = '')
    client.close()
    print()
    
```
